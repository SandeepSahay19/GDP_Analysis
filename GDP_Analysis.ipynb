```python
import pandas as pd
df= pd.read_csv("India_GDP(1960-2022).csv")
import matplotlib.pyplot as plt
```

```python
df
```

# ABOUT THE PROBLEM:

## I have been provided with the dataset of INDIA GDP (1960 - 2022).I have to find Insights and trends formed over the years and answer questions below mentioned :

## 1. What is the trend in GDP (in billion dollars) over the years ?

```python
gdp_by_year = df.groupby("Year")["GDP in (Billion) $"].mean()
plt.figure(figsize=(10, 6))
plt.plot(gdp_by_year.index, gdp_by_year.values,marker='o')
plt.xlabel("Year")
plt.ylabel("GDP (billion dollars)")
plt.title("Trend in GDP over the Years")
plt.grid(True)
plt.show()
```

## 2. What is the trend in per capita income (rupees) over the years ?

```python
per_capita_income_by_year = df.groupby("Year")["Per Capita in rupees"].mean()
plt.figure(figsize=(10, 6))
plt.plot(per_capita_income_by_year.index, per_capita_income_by_year.values, marker ='o')
plt.xlabel("Year")
plt.ylabel("Per Capita Income (rupees)")
plt.title("Changes in Per Capita Income over the Years")
plt.grid(True)
plt.show()
```

## 3.Year with the highest GDP and it's Corresponding growth rate?

```python
row_highest_gdp = df[df["GDP in (Billion) $"] == df["GDP in (Billion) $"].max()]

# Extract the year and growth rate from the row with the highest GDP
year_highest_gdp = row_highest_gdp["Year"].values[0]
growth_rate_highest_gdp = row_highest_gdp["Growth %"].values[0]

# Print the result
print("Year with the highest GDP: ", year_highest_gdp)
print("Corresponding growth rate: ", growth_rate_highest_gdp, "%")
```

## 4.Years with the negative Growth rate percentage?

```python
negative_growth_years = df[df["Growth %"] < 0]

# Extract the years with negative growth rates
years_with_negative_growth = negative_growth_years["Year"].values

# Print the years with negative growth rates
if len(years_with_negative_growth) > 0:
    print("Years with negative growth rates:")
    for year in years_with_negative_growth:
        print(year)
else:
    print("No years with negative growth rates found in the dataset.")
```

## 5.Average growth rate of the country ?

```python
average_growth_rate = df["Growth %"].mean()

# Print the average growth rate of GDP
print("Average growth rate of GDP: {:.2f}%".format(average_growth_rate))
```

## 6. Growth Rate of GDP over the years ?

```python
plt.figure(figsize=(10, 6))  # Optional: adjust the size of the plot
plt.plot(df["Year"], df["Growth %"], marker='o')  # Line plot with markers for data points
plt.title("Growth Rate of GDP over the Years")
plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.grid(True)
plt.show()
```

## 7.Top five Years with the most Per Capita in rupees ?

```python
df_sorted = df.sort_values(by="Per Capita in rupees", ascending=False)

# Select the top five rows with the highest per capita income in rupees
top_five_years = df_sorted.head(5)

# Print the years with the highest per capita income in rupees
print("Top five years with highest per capita income in rupees:")
print(top_five_years[["Year", "Per Capita in rupees"]])
```

```python
plt.figure(figsize=(10, 6)) 
plt.bar(top_five_years["Year"], top_five_years["Per Capita in rupees"])  # Bar graph with Year on x-axis and Per_capita_rupees on y-axis
plt.title("Top Five Years with Highest Per Capita Income in Rupees")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (Rupees)")
plt.grid(True)
plt.show()
```

## 8. What is the percentage change in per capita income (in rupees) from one year to the next? Which year had the highest percentage increase and which year had the highest percentage decrease?

```python
df_sorted = df.sort_values(by="Year", ascending=True)

# Calculate the percentage change in Per_capita_rupees column from one year to the next
df_sorted["Per_capita_rupees_percentage_change"] = df_sorted["Per Capita in rupees"].pct_change() * 100

# Find the year with the highest percentage increase and highest percentage decrease
year_highest_increase = df_sorted.loc[df_sorted["Per_capita_rupees_percentage_change"].idxmax(), "Year"]
highest_increase = df_sorted["Per_capita_rupees_percentage_change"].max()
year_highest_decrease = df_sorted.loc[df_sorted["Per_capita_rupees_percentage_change"].idxmin(), "Year"]
highest_decrease = df_sorted["Per_capita_rupees_percentage_change"].min()

# Print the results
print("Year with highest percentage increase in per capita income (in rupees):", year_highest_increase)
print("Percentage increase:", highest_increase)
print("Year with highest percentage decrease in per capita income (in rupees):", year_highest_decrease)
print("Percentage decrease:", highest_decrease)
```

## 9.Top five years with highest Growth percentage of GDP over the years ?

```python
sorted_df = df.sort_values(by="Growth %", ascending=False)

# Select the top five years with the highest growth rate
top_five_years = sorted_df.head(5)

# Plot a bar chart for the top five years with the highest growth rate
plt.figure(figsize=(10, 6))
plt.bar(top_five_years["Year"], top_five_years["Growth %"])
plt.xlabel("Year")
plt.ylabel("Growth %")
plt.title("Top Five Years with Highest Growth Rate")
plt.grid(True)
plt.show()

```

## 10.Top five years with lowest Growth percentage of GDP over the years ?

```python
sorted_df = df.sort_values(by="Growth %", ascending=True)

# Select the top five years with the lowest growth rate
top_five_years = sorted_df.head(5)

# Plot a bar chart for the top five years with the lowest growth rate
plt.figure(figsize=(10, 6))
plt.bar(top_five_years["Year"], top_five_years["Growth %"])
plt.xlabel("Year")
plt.ylabel("Growth %")
plt.title("Top Five Years with Lowest Growth Rate")
plt.show()
```
